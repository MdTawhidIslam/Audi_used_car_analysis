```python
fuelType_count = df['fuelType'].value_counts(normalize=True)
fuelType_count = pd.DataFrame(fuelType_count)

fuelType_count = fuelType_count.reset_index()

display(fuelType_count)

fuelType_count = fuelType_count.rename(columns={'index':'fuelType', 'fuelType':'% of cars'})
display(fuelType_count)

fuelType_count['% of cars'] = fuelType_count['% of cars']*100
fuelType_count = fuelType_count.round(2)
display(fuelType_count)
```

# Barplot

```python
sns.barplot(x='fuelType', y='% of cars', data=fuelType_count)

plt.title("Percentage of AUDI Cars")



plt.show()
```

```python
transmission_count = df['transmission'].value_counts(normalize=True)
transmission_count = pd.DataFrame(transmission_count)

transmission_count = transmission_count.reset_index()

display(transmission_count)

transmission_count = transmission_count.rename(columns={'index':'transmission', 'transmission':'% of cars'})
display(transmission_count)

transmission_count['% of cars'] = transmission_count['% of cars']*100
transmission_count = transmission_count.round(2)
display(transmission_count)
```

```python
sns.barplot(x='transmission', y='% of cars', data=transmission_count)

plt.title("Percentage of AUDI Cars")



plt.show()
```

```python
model_count = df['model'].value_counts(normalize=True)
model_count = pd.DataFrame(model_count)

model_count = model_count.reset_index()

display(model_count)

model_count = model_count.rename(columns={'index':'model', 'model':'% of cars'})
display(model_count)

model_count['% of cars'] = model_count['% of cars']*100
model_count = model_count.round(2)
display(model_count)
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x='model', y='% of cars', data=model_count, color='blue')

plt.title("Percentage of AUDI Cars")


plt.yticks(np.arange(0,21,2))
plt.show()
```

# Function Creation

```python
def unique_percent_count(data, col):

    col_count = data[col].value_counts(normalize=True)
    col_count = pd.DataFrame(col_count)

    col_count = col_count.reset_index()

    

    col_count = col_count.rename(columns={'index':col, col:'% of cars'})
    col_count['% of cars'] = col_count['% of cars']*100
    

    col_count = col_count.round(2)
    
    return col_count

```

```python
fuelType_count = unique_percent_count(df, 'fuelType')

display(fuelType_count)
```

```python
transmission_count = unique_percent_count(df, 'transmission')

display(transmission_count)

    
```

```python
model_count = unique_percent_count(df, 'model')

display(model_count)
```

```python
year_count = unique_percent_count(df, 'year')

#display(year_count)

year_count = year_count.sort_values(by='year')

display(year_count.head(5))
display(year_count.tail(5))

```

# Create another FUNCTION for the visualization

```python
def barplot(data, x_col, y_col, color, rotation):
    sns.barplot(x=x_col, y=y_col, data=data, color=color)
    plt.title("Percentage of AUDI Cars for each " + x_col )
    
    #plt.yticks(np.arange(0,21,2))
    plt.xticks(rotation=rotation)
```

```python
sns.set_context("talk")

plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
barplot(model_count, 'model', '% of cars', 'blue', 90)
plt.yticks(np.arange(0,21,2))



plt.subplot(2,2,2)
barplot(transmission_count, 'transmission', '% of cars', 'orange', 0)




plt.subplot(2,2,3)
barplot(fuelType_count, 'fuelType', '% of cars', 'green', 0)


plt.subplot(2,2,4)
barplot(year_count, 'year', '% of cars', 'yellow', 90)



plt.suptitle("Categorical variable in the AUDI used Cars")


plt.tight_layout()
#plt.savefig("barplot.png")

plt.savefig("barplot2.jpg")

plt.show()
```

# ScatterPlot 

```python
plt.figure(figsize=(20,12))

# mileage vs price
plt.subplot(2,2,1)
sns.scatterplot(x='mileage', y='price', data=df)

# mpg vs price
plt.subplot(2,2,2)
sns.scatterplot(x='mpg', y='price', data=df)

# engineSize vs price
plt.subplot(2,2,3)
sns.scatterplot(x='engineSize', y='price', data=df)

# mpg vs enginesize
plt.subplot(2,2,4)
sns.scatterplot(x='mpg', y='engineSize', data=df)

plt.tight_layout()
plt.show()
```

# Scatterplot with trend line

```python
plt.figure(figsize=(20,12))
plt.subplot(3,1,1)
sns.regplot(x='mileage', y='price', data=df)


plt.subplot(3,1,2)
sns.regplot(x='mileage', y='price', data=df, line_kws={'color':'r'})


plt.subplot(3,1,3)
sns.regplot(x='mileage', y='price', data=df, 
            scatter_kws={'color':'orange', 'edgecolor':'white'})

plt.tight_layout()
plt.show()
```

# Histogram 

```python
plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
sns.histplot(data=df, x='mpg')
#plt.xticks(np.arange(0,151,20))
plt.title("Audi used car's mpg")

plt.subplot(2,2,2)
sns.histplot(data=df, x='mpg', binwidth=5)
#plt.xticks(np.arange(0,151,20))
plt.title("Audi used car's mpg")

plt.subplot(2,2,3)
sns.histplot(data=df, x='mpg', binwidth=10, color='red')
#plt.xticks(np.arange(0,151,20))
plt.title("Audi used car's mpg")

plt.subplot(2,2,4)
sns.histplot(data=df, x='mpg', binwidth=10, color='green', fill=False)
plt.title("Audi used car's mpg")




plt.tight_layout()
plt.show()
```

# Distribution  & ECDF Plot

```python
plt.figure(figsize=(20,18))

plt.subplot(3,2,1)
sns.histplot(data=df, x='mpg', binwidth=1)
plt.title("row = 3, column = 3, position = 1")

plt.subplot(3,2,2)
sns.histplot(data=df, x='mpg', binwidth=1, cumulative=True)
plt.grid()
plt.title("row = 3, column = 2, position = 2")

plt.subplot(3,2,3)
sns.histplot(data=df, x='mpg', binwidth=1, kde=True)
plt.title("row = 3, column = 2, position = 3")

plt.subplot(3,2,4)
sns.histplot(data=df, x='mpg', binwidth=1, kde=True, hue='fuelType')
plt.title("row = 3, column = 2, position = 4")

plt.subplot(3,2,5)
sns.histplot(data=df, x='mpg', binwidth=1, cumulative=True, stat='probability', hue='fuelType')
plt.grid()
plt.title("row = 3, column = 2, position = 5")

plt.tight_layout()
plt.show()
```
